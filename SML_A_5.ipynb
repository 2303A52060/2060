```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load the data from the CSV file
data = pd.read_csv('Salary_Data.csv')

# Step 2: Display the first few rows of the dataset to understand its structure
print("First few rows of the dataset:")
print(data.head())

# Step 3: List all column names to identify features and target variables
print("\nColumn names in the dataset:")
print(data.columns)

# Assume 'YearsExperience' is the feature and 'Salary' is the target variable
# Step 4: Plot a graph between the feature and the target variable
plt.figure(figsize=(8, 6))
plt.scatter(data['YearsExperience'], data['Salary'], color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Years of Experience vs Salary')
plt.grid(True)
plt.show()

# Step 5: Normalize the features and target variable
scaler = MinMaxScaler()

# Normalize 'YearsExperience'
data[['YearsExperience']] = scaler.fit_transform(data[['YearsExperience']])

# Normalize 'Salary'
data[['Salary']] = scaler.fit_transform(data[['Salary']])

# Display the first few rows of the normalized dataset
print("\nFirst few rows of the normalized dataset:")
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Step 1: Read the Data
df = pd.read_csv('Salary_Data.csv')

# Display the first few rows of the DataFrame
print("Initial Data:")
print(df.head())

# Step 2: Find Features and Target Variables
# Assuming 'Salary' is the target variable
target = df['Salary']
features = df.drop(columns=['Salary'])

# Display feature columns and target
print("\nFeatures:")
print(features.head())
print("\nTarget:")
print(target.head())

# Step 3: Plot Graphs
for column in features.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(features[column], target, alpha=0.5)
    plt.title(f'{column} vs Salary')
    plt.xlabel(column)
    plt.ylabel('Salary')
    plt.grid(True)
    plt.show()

# Step 4: Normalize the Data (0 to 1)
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(features)

# Convert normalized features back to a DataFrame
normalized_features_df = pd.DataFrame(normalized_features, columns=features.columns)

# Display the first few rows of the normalized features
print("\nNormalized Features:")
print(normalized_features_df.head())

# Step 5: Convert String-Valued Features to Numbers
label_encoders = {}
for column in features.columns:
    if features[column].dtype == 'object':  # Check if the column is of object type
        le = LabelEncoder()
        features[column] = le.fit_transform(features[column])
        label_encoders[column] = le  # Save the encoder for future use

# Display the first few rows of the features with converted values
print("\nFeatures with Encoded Strings:")
print(features.head())
```
