From the above data:

Read the data with pandas and find features and target variables

Plot a graph between features and target

Find Best fit line using linear regression.

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Read the data
data = pd.read_csv("Salary_Data.csv")

# Step 2: Extract features and target
X = data[['YearsExperience']]
y = data['Salary']

# Step 3: Create a linear regression model and fit it
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict values using the model (best fit line)
y_pred = model.predict(X)

# Step 5: Plot the graph
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Best fit line')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

Read the data with pandas and find features and target variables

Plot a graph between features and target

Find Best fit line using linear regression.

Feature 1: 1.2 2.4 3.5 4.1 1.9 3.7 2.1 3.3 2.8 4

Feature 2: 2.3 1.9 2.7 3 2.8 2.5 3.2 2.4 3.1 3.3

Feature 3: 3.1 2.8 1.5 3.6 2.5 1.9 2.2 3 1.8 2.7

Feature 4: 4.2 3.5 2.9 4.8 3.2 4 4.1 4.5 3.6 4.9

Target (Y):15.6 13.1 12.4 18.2 14.1 16 14.5 17.3 14.9 19.2








```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'Feature1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
    'Feature2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
    'Feature3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
    'Feature4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
    'Target': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
}

df = pd.DataFrame(data)


features = ['Feature1', 'Feature2', 'Feature3', 'Feature4']

plt.figure(figsize=(14, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    plt.scatter(df[feature], df['Target'], color='blue', label='Data Points')


    X = df[[feature]]
    y = df['Target']
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)


    plt.plot(df[feature], y_pred, color='red', linewidth=2, label='Best Fit Line')
    plt.xlabel(feature)
    plt.ylabel('Target')
    plt.title(f'{feature} vs Target')
    plt.legend()

plt.tight_layout()
plt.show()

for feature in features:
    X = df[[feature]]
    y = df['Target']
    model = LinearRegression()
    model.fit(X, y)
    print(f'{feature} - Coefficient (Slope): {model.coef_[0]}, Intercept: {model.intercept_}')
```
