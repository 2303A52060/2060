Given the following dataset: X={1.3,1.5,2.6,2.6,3.2,3.9,4.2,3.7,3.10,3.0,11,6.7,1.9}. Compute the density of the data using a histogram with 5 bins.

Question 3:

From the above data Find Mean, Median, Mode, Variance, Standard deviation, Range , Interquartile Range (IQR),  Skewness ,Kurtosis of each feature.


```python
import numpy as np
import pandas as pd
from scipy import stats

# Dataset for Question 1
X = np.array([1.3, 1.5, 2.6, 2.6, 3.2, 3.9, 4.2, 3.7, 3.10, 3.0, 11, 6.7, 1.9])

# Define bins
num_bins = 5
bin_edges = np.linspace(X.min(), X.max(), num_bins + 1)

# Compute histogram
hist, bin_edges = np.histogram(X, bins=bin_edges)
bin_width = bin_edges[1] - bin_edges[0]
density = hist / (bin_width * len(X))

# Print density results
print("Question 1: Histogram Density")
for i in range(num_bins):
    print(f"Bin {i + 1}: {bin_edges[i]:.2f} to {bin_edges[i + 1]:.2f}, Density: {density[i]:.4f}")

# Data for Question 2
dependent = np.array([12, 13, 56, 23, 41, 28, 51])
independent = np.array([51, 63, 37, 88, 33, 72, 81])

# Function to compute statistics
def compute_statistics(data):
    mean = np.mean(data)
    median = np.median(data)
    mode_result = stats.mode(data)
    mode = mode_result.mode  # Extract the mode value
    variance = np.var(data, ddof=0)
    std_dev = np.std(data, ddof=0)
    data_range = np.ptp(data)
    iqr = stats.iqr(data)
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    return {
        'Mean': mean,
        'Median': median,
        'Mode': mode,
        'Variance': variance,
        'Standard Deviation': std_dev,
        'Range': data_range,
        'IQR': iqr,
        'Skewness': skewness,
        'Kurtosis': kurtosis
    }

# Compute statistics for dependent and independent variables
stats_dependent = compute_statistics(dependent)
stats_independent = compute_statistics(independent)

# Print results for Question 2
print("\nQuestion 2: Statistics for Dependent Variable")
for stat_name, value in stats_dependent.items():
    print(f"{stat_name}: {value:.4f}")

print("\nQuestion 2: Statistics for Independent Variable")
for stat_name, value in stats_independent.items():
    print(f"{stat_name}: {value:.4f}")

# Data for Question 3 (same as above data)
# For each feature in the dataset, the statistics computed are from Question 2
# We already have these statistics computed.

print("\nQuestion 3: Statistics for Each Feature (same as above data)")
print("Dependent Variable Statistics:")
for stat_name, value in stats_dependent.items():
    print(f"{stat_name}: {value:.4f}")

print("\nIndependent Variable Statistics:")
for stat_name, value in stats_independent.items():
    print(f"{stat_name}: {value:.4f}")

```
